```python
#【問題1】乱数の作成
import numpy as np
from matplotlib import pyplot as plt

mean  = np.array([-3, 0])
cov = np.array([[1.0, 0.8], 
                            [0.8, 1.0]])

np.random.seed(0)
random_data = np.random.multivariate_normal(mean, cov, size=500)
print(random_data.shape)
```

```python
#【問題2】散布図による可視化
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(0)

mean  = np.array([-3, 0])
cov = np.array([[1.0, 0.8], 
                            [0.8, 1.0]])

random_data = np.random.multivariate_normal(mean, cov, size=500)

plt.scatter(random_data[0:,0], random_data[0:,1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.show()
```

```python
#【問題3】ヒストグラムによる可視化
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(0)

mean  = np.array([-3, 0])
cov = np.array([[1.0, 0.8], 
                            [0.8, 1.0]])

random_data = np.random.multivariate_normal(mean, cov, size=500)

plt.hist(random_data[0:,0], bins = 100)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")
plt.xlim(-6, 4)
plt.show()

plt.hist(random_data[0:,1], bins = 100)
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.xlim(-6, 4)
plt.show()
```

```python
#【問題4】データの追加
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(0)

mean_0 = np.array([-3, 0])
cov_0 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

mean_1 = np.array([0, -3])
cov_1 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

random_data_0 = np.random.multivariate_normal(mean_0, cov_0, size=500)
random_data_1 = np.random.multivariate_normal(mean_1, cov_1, size=500)

plt.scatter(random_data_0[0:,0], random_data_0[0:,1], label='0')
plt.scatter(random_data_1[0:,0], random_data_1[0:,1], label='1')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.legend()
plt.show()
```

```python
#【問題5】データの結合
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(0)

mean_0 = np.array([-3, 0])
cov_0 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

mean_1 = np.array([0, -3])
cov_1 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

random_data_0 = np.random.multivariate_normal(mean_0, cov_0, size=500)
random_data_1 = np.random.multivariate_normal(mean_1, cov_1, size=500)

print(np.vstack( (random_data_0, random_data_1) ).shape)
```

```python
#【問題6】ラベル付け
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(0)

mean_0 = np.array([-3, 0])
cov_0 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

mean_1 = np.array([0, -3])
cov_1 = np.array([[1.0, 0.8], 
                                 [0.8, 1.0]])

random_data_0 = np.random.multivariate_normal(mean_0, cov_0, size=500)
random_data_0 = np.hstack((random_data_0, np.zeros(500).reshape(500, 1)))

random_data_1 = np.random.multivariate_normal(mean_1, cov_1, size=500)
random_data_1 = np.hstack((random_data_1, np.ones(500).reshape(500, 1)))

print(np.vstack( (random_data_0, random_data_1) ).shape)
print(np.vstack( (random_data_0, random_data_1) ))
```
