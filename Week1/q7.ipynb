```python
"""
自身にて改訂。
"""
import matplotlib.pyplot as plt
%matplotlib inline

THICKNESS = 0.00008/1000
folding_num = 43

folded_thickness = THICKNESS
folded_thickness_list = []
folded_thickness_list.append(THICKNESS)

#繰り返し計算
for i in range(folding_num):
    folded_thickness *= 2
    folded_thickness_list.append(folded_thickness)

#グラフ描画
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[km]")
plt.tick_params(labelsize=20)
plt.hlines([0, 200000, 400000, 600000, 800000], -5, 45, "blue", linestyles='dashed', alpha=0.5)
plt.hlines([384400], -5, 45, "red", linestyles='dashed', alpha=1, label='Distance to the moon')
plt.vlines([0, 20, 40], -10, 810000, "blue", linestyles='dashed', alpha=0.5)
plt.plot(folded_thickness_list, color='m', marker='o')
plt.legend()
plt.show()
```
