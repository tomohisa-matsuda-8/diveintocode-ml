```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

""""
宇宙の大きさはWikipediaの記載(下記)より,3*10**80立方メートルとする。
https://ja.wikipedia.org/wiki/%E8%A6%B3%E6%B8%AC%E5%8F%AF%E8%83%BD%E3%81%AA%E5%AE%87%E5%AE%99
饅頭の大きさは、一般的なサイズとして、0.0001立方メートルとする。
"""

VOL_SPACE = 3*10**80
VOL_INITIAL_MANJU = 0.0001

def calc_manju_size(vol_manju, vol_space):
    vol_manju_list = []
    vol_manju_list.append(vol_manju)

    #饅頭は5分で２倍になるということは、１時間で、2**12倍となる。
    #本プラグラムでは、饅頭サイズ > 宇宙の体積となる経過時間(hour, 変数:countに格納)を求めている。
    count = 0
    while vol_manju_list[-1] < vol_space:
        vol_manju = vol_manju * (2**12)
        vol_manju_list.append(vol_manju)
        count += 1
    return count, vol_manju_list

#変数:over hourに饅頭サイズ > 宇宙の体積となる経過時間(hour)が返される。
over_hour, manju_size_list = calc_manju_size(vol_manju=VOL_INITIAL_MANJU, vol_space=VOL_SPACE)
print("{}時間後に饅頭のサイズは宇宙を超えます".format(over_hour))

#グラフにプロット
plt.title("Volume of the manju")
plt.xlabel("elapsed_hour")
plt.ylabel("volume[m3]")
plt.plot(manju_size_list, marker='o', label='Volume of manju') 
plt.hlines([3*10**80], 0, 25, "red", linestyles='dashed', alpha=1, label='Volume of space')
plt.legend()
plt.show()

#グラフにプロット(manju曲線の立ち上がりを確認し易くするために縦軸を調整した。)
plt.title("Volume of the manju")
plt.xlabel("elapsed　hour")
plt.ylabel("volume[m3]")
plt.plot(manju_size_list, marker='o', label='Volume of manju') 
plt.hlines([3*10**80], 0, 25, "red", linestyles='dashed', alpha=1, label='Volume of space')
plt.ylim([0,3.1e80])
plt.legend()
plt.show()
```
