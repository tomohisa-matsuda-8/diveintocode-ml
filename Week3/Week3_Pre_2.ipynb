### 【問題4】欠損値への対応

欠損値に関して以下のことを行い、レポートしてください。


各特徴量の欠損値の有無を確認する。

```python
import missingno as msno
%matplotlib inline
print(msno.matrix(df))
```

欠損値が何パーセント含まれるかを確認する。

```python
array_count = df.count().values
array_missing_rate = array_count / 1460
array = np.concatenate((array_count[:, np.newaxis], array_missing_rate[:, np.newaxis]), axis=1) 
array_label_column = df.columns
array_label_index = df.index

missing_table = pd.DataFrame(array)
missing_table.index = array_label_column
missing_table.columns = ['total', 'missing_ratio']
missing_table
```

5つ以上欠損値がある特徴量（列）は削除する。

```python
array_count = df.count().values
drop_columns_list = []
for i in range(81):
    if (1460 - array_count[i]) >= 5:     #5つ以上欠損値がある特徴量（列）の判定
        drop_columns_list.append(i)

df_temp = df.drop(df.columns[drop_columns_list], axis=1)
df_temp
```

欠損値があるサンプル（行）は削除する。

```python
array_count = df_temp.count(axis=1).values  #df_temp:列削除済みのデータフレーム
drop_index_list = []
for i in range(1460):
    if (63 - array_count[i]) >= 1:     #1つ以上欠損値がある行の判定
        drop_index_list.append(i)

df_dropped = df_temp.drop(df_temp.index[drop_index_list], axis=0)
df_dropped
```

### 【問題6】分布の確認
seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

目的変数に対して対数変換を行う。

対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
from scipy.stats import kurtosis
from scipy.stats import skew

X = df_dropped.loc[:, 'SalePrice']
sns.distplot(X, kde=False, rug=False, bins=50) 
plt.show()
print("対数変換前の尖度：{0:.2f}".format(kurtosis(X)))
print("対数変換前の歪度：{0:.2f}".format(skew(X)))

X_log = X.apply(np.log)
sns.distplot(X_log, kde=False, rug=False, bins=50) 
plt.show()
print("対数変換後の尖度：{0:.2f}".format(kurtosis(X_log)))
print("対数変換後の歪度：{0:.2f}".format(skew(X_log)))
```

歪度が正の値をとり、尖度が大きく、数値が大きい方に裾を引く分布は、

観測値を対数変換すると正規分布に近くなる。このような分布を対数正規分布と呼ぶ。

今回の目的変数は対数正規分布であると考えられるため、対数変換して正規分布に変換すると、尖度、歪度ともに小さくなる。

### 【問題7】相関係数の確認
特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。

相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
corr_mat = df_dropped.corr( method='pearson' )
sns.set()
sns.set_context("notebook")
plt.figure(figsize=(20, 10))

sns.heatmap(
    corr_mat,
    annot = True,
    xticklabels = corr_mat.columns.values,
    yticklabels = corr_mat.columns.values,
)
plt.show()
```
